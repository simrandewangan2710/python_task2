```python
#List Append Example
#The append() method appends(Add) an element to the end of the list.
vegetables = ["potato", "tomato", "cabbage"]
vegetables.append("onion")
print(vegetables)
```

```python
#List Clear Example
#The clear() method removes all the elements from a list.
vegetables = ["potato", "tomato", "cabbage"]
vegetables.clear()
print(vegetables)
```

```python
#List Copy Example
#The copy() method returns a copy of the specified list.
vegetables = ["potato", "tomato", "cabbage"]
x = vegetables.copy()
print(x)
```

```python
#List Extend Example
#The extend() method adds the specified list elements (or any iterable) to the end of the current list.
vegetables = ["potato", "tomato", "cabbage"]    
fruits = ["apple", "banana", "cherry"]
vegetables.extend(fruits)
print(vegetables)
```

```python
#List insert Example
#The insert() method inserts the specified value at the specified position.
vegetables = ["potato", "tomato", "cabbage"]
vegetables.insert(1, "onion")
print(vegetables)
```

```python
#List Remove Example
#The remove() method removes the specified item from the list.
vegetables = ["potato", "tomato", "cabbage"]
vegetables.remove("tomato")
print(vegetables)
```

```python
#List Reverse Example
#The reverse() method reverses the sorting order of the elements.   
vegetables = ["potato", "tomato", "cabbage"]
vegetables.reverse()
print(vegetables)
```

```python
#List Sort Example
#The sort() method sorts the list ascending by default.
vegetables = ["potato", "tomato", "cabbage"]
vegetables.sort()
print(vegetables)
```

```python
#List Pop Example
#The pop() method removes the specified index, (or the last item if index is not specified).
vegetables = ["potato", "tomato", "cabbage"]
vegetables.pop(0)
print(vegetables)
```

```python
#List Count Example
#The count() method returns the number of elements with the specified value.
vegetables = ["potato", "tomato", "cabbage", "tomato"]
x = vegetables.count("tomato")
print(x)
```
