```python
#string capitalize() method
#The capitalize() method returns a string where the first character is upper case.
txt = "hello, my name is simran."
x = txt.capitalize()
print (x)
```

```python
#string casefold() method
#The casefold() method returns a string where all characters are lower case.
txt = "Hello, My Name Is Simran!"
x = txt.casefold()
print(x)
```

```python
#string center() method
#The center() method will center align the string, using a specified character (space is default) as the fill character.
txt = "simran"
x = txt.center(20)
print(x)
```

```python
#string count() method
#The count() method returns the number of times a specified value occurs in a string.
txt = "I love pineapples, pineapple are my favorite fruit"
x = txt.count("pineapple")
print(x)
```

```python
#string encode() method
#The encode() method encodes the string, using the specified encoding. If no encoding is specified, UTF-8 will be used.
txt = "My name is Simran"
x = txt.encode()
print(x)
```

```python
#string endswith() method   
#The endswith() method returns True if the string ends with the specified value, otherwise False.
txt = "Hello, welcome to vscode notebook."
x = txt.endswith(".")
print(x)
```

```python
#string find() method   
#The find() method finds the first occurrence of the specified value. If the value is not found, it returns -1.
txt = "Hello, welcome to vscode notebook."
x = txt.find("vscode")  
print(x)
```

```python
#String translate() method
#The translate() method returns a string where some specified characters are replaced with the character described in a dictionary, or in a mapping table.
txt = "Hello my friends"
mytable = str.maketrans("H", "J")
x = txt.translate(mytable)
print(x)
```

```python
#String islower() method
#The islower() method returns True if all characters in the string are lower case.
txt = "hello simran!"
x = txt.islower()
print(x)
```

```python
#String isupper() method
#The isupper() method returns True if all characters in the string are upper case.
txt = "HELLO SIMRAN!"
x = txt.isupper()
print(x)
```

```python
#String Join() method
#The join() method takes all items in an iterable and joins them into one string.
mytuple = ("Simran", "Kaur", "Singh")
x = " ".join(mytuple)
print(x)
```

```python
#String Strip() method
#The strip() method removes any leading (spaces at the beginning) and trailing (spaces at the end) characters (space is the default leading character to remove)
txt = "   cherry  "
x = txt.strip()
print(x)
print("of all fruits", x, "is my favorite")
```

```python
#String Title() method
#The title() method returns a string where the first character in every word is upper case.
txt = "welcome to vscode notebook"
x = txt.title()
print(x)
```

```python
#String Replace() method
#The replace() method replaces a specified phrase with another specified phrase.
txt = "I like junk food"
x = txt.replace("junk food", "healthy food")
print(x)
```
